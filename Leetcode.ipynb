### 26. Remove Duplicates from sorted array

```
"Start at index 1 and compare to index 1-1 =0. If the two consequtive elements do not equal each other that satisfies our condition. So, we copy that to the array."
```


````python
from typing import List
class Solution:
  def removeDuplicates(self, nums: List[int]) -> int:
    k=1
    for i in range(1, len(nums)):
      if nums[i] != nums[k-1]:
        nums[k] = nums[i]
        k += 1
    return k

if __name__ == '__main__':
  solution = Solution()
  unique_count = solution.removeDuplicates([1,2,3,4,4])
  print(unique_count)
````
