### EASY 26. Remove Duplicates from sorted array - Array/String
Start at index 1 and compare to index 1-1 =0. If the two consequtive elements do not equal each other that satisfies our condition. So, we copy that to the array.


```python
from typing import List
class Solution:
  def removeDuplicates(self, nums: List[int]) -> int:
    k=1
    for i in range(1, len(nums)):
      if nums[i] != nums[k-1]:
        nums[k] = nums[i]
        k += 1
    return k

if __name__ == '__main__':
  solution = Solution()
  remove_duplicate = solution.removeDuplicates([1,2,3,4,4])
  print(remove_duplicate)
```

### EASY Two Sum - Array/String

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i, num in enumerate(nums):
            looking_for = target - num
            if looking_for in seen:
                return [seen[looking_for], i]

            seen[num] = i

if __name__ == '__main__':
  solution = Solution()
  two_sum = solution.twoSum([2,7,11,15], 9)
  print(two_sum)
```

### EASY Contains Duplicate - Array/Hashing
create set, iterate through the list, if it's in the set we made, then return true, otherwise add it to the set and continue

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for num in nums:
            if num in seen:
                return True
            seen.add(num)
        return False

if __name__ == '__main__':
  solution = Solution()
  contains_duplicate = solution.containsDuplicate([1,2,3,1])
  print(contains_duplicate)
```

### EASY - Valid Anagram
if the values are not equal in size, they can't be anagrams, so don't check first, as sorted for the length can be inefficient. sorted and check if equal.

```python
class Solution:
  def isAnagram(self, s: str, t: str) -> bool:
          if len(s) != len(t):
              return False
          return sorted(s) == sorted(t)
if __name__ == '__main__':
  solution = Solution()
  is_anagram = solution.isAnagram(s = "anagram", t = "nagaram")
  print(is_anagram)
```

### EASY - Valid Palindrome
regex to remove special characters, then .lower to change str to lower case. We reverse the string, and return True if they match.

```python
import re
class Solution:
    def isPalindrome(self, s: str) -> bool:
        phrase = (re.sub('[^a-zA-Z0-9]','', s)).lower()
        reversed_phrase = ''.join(reversed(phrase))
        return phrase == reversed_phrase
if __name__ == '__main__':
  solution = Solution()
  is_palindrome = solution.isPalindrome("A man, a plan, a canal: Panama")
  print(is_palindrome)
```

### Valid Parentheses - EASY
for the first iteration we will add the bracket to the stack, then, we compare the next character to that bracket, if it matches the key value pair, then we remove the bracket from the stack and continue otherwise we return False and the method returns False.

```python
class Solution:
    def isValid(self, s: str) -> bool:
      stack = []
      is_valid_pair = {"}":"{", "]":"[", ")":"("}
      for char in s:
        if char in is_valid_pair:
          if stack and stack[-1] == is_valid_pair[char]:
            stack.pop()
          else:
            return False
        else:
          stack.append(char)
      return True if not stack else False

if __name__ == '__main__':
  solution = Solution()
  is_valid = solution.isValid("([])")
  print(is_valid)
```
